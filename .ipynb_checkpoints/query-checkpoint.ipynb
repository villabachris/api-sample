```python
def get_updated_customers():
    sendah_reg = pd.read_csv('sendah_customers_reg.csv')
    sendah_reg['id_customer'] = sendah_reg['id_customer'].str.replace(',', '').astype(int)
    sendah_reg['date_add'] = pd.to_datetime(sendah_reg['date_add'])
    sendah_reg['full_name'] = sendah_reg['firstname'].str.title() + ' ' + sendah_reg['lastname'].str.title()
    sendah_reg = sendah_reg[['id_customer', 'date_add', 'full_name']]
    
    query = '''select 
id_customer,
firstname,
lastname,
date_add
from sendahdb.sd_customer
where id_customer>{}'''.format(sendah_reg.tail(1)['id_customer'].values[0])

    mycursor = mydb.cursor()
    mycursor.execute(query)
    myresult = pd.DataFrame(mycursor.fetchall())
    
    if myresult.shape[0] != 0:
        myresult.columns = ['id_customer', 'firstname', 'lastname', 'date_add']
        myresult['date_add'] = pd.to_datetime(myresult['date_add'])
        myresult['full_name'] = myresult['firstname'].str.title() + ' ' + myresult['lastname'].str.title()
        return pd.concat([sendah_reg, myresult], axis=0, ignore_index=True).set_index('id_customer')
    else:
        return sendah_reg.set_index('id_customer')
```

```python
def get_shopsend_transactions(year, month):
    query = """SELECT
pl.name,
o.date_upd,
c.email,
c.id_customer,
round((od.product_price * od.product_quantity) - if(od.reduction_percent, (od.product_price * od.product_quantity) * od.reduction_percent / 100, od.reduction_amount), 2),
od.product_quantity AS Qty
FROM sd_orders AS o
JOIN sd_order_detail AS od ON od.id_order = o.id_order
JOIN sd_product AS p ON p.id_product = od.product_id
JOIN sd_product_lang AS pl ON pl.id_product = p.id_product
JOIN sd_order_state AS os ON os.id_order_state = o.current_state
JOIN sd_order_state_lang AS osl ON osl.id_order_state = os.id_order_state
JOIN sd_customer AS c ON c.id_customer = o.id_customer
WHERE
MONTH(o.date_upd) = {} and YEAR(o.date_upd) = {}
AND
osl.name IN ('Delivered', 'Shipped', 'Order Fulfilled')
ORDER BY
  o.date_upd ASC""".format(month, year)

    mycursor = mydb.cursor()
    mycursor.execute(query)
    myresult = pd.DataFrame(mycursor.fetchall())
    if myresult.shape[0] != 0:
        myresult.columns = ['item', 'delivery_date', 'customer_id', 'shop_id', 'amount', 'quantity']
        myresult['amount'] = myresult['amount'].astype(float)
        myresult['type'] = 'shop'
        return myresult[['item', 'delivery_date', 'customer_id', 'shop_id', 'amount', 'quantity', 'type']]
    else:
        return pd.DataFrame()
```

```python
def get_num_customers(df):
    return len(df['customer_id'].unique())
```

```python
def get_num_transactions(df):
    return int(df.shape[0])
```

```python
def get_sales_amount(df):
    # get total sales
    return float(df['amount'].sum())
```

```python
def get_top_n_customers_by_amount(transactions, customers, n=10):
    if n == None:
        return transactions.groupby('customer_id')['amount'].sum().sort_values(ascending=False)
    grouped = transactions.groupby('customer_id')['amount'].sum().sort_values(ascending=False).head(n)
#     grouped.index = [customers.loc[x, 'full_name'] if x in customers.index else x for x in grouped.index]
    return grouped
```

```python
def get_top_n_items_by_amount(transactions, n=10):
    if n == None:
        return transactions.groupby('item')['amount'].sum().sort_values(ascending=False)
    return transactions.groupby('item')['amount'].sum().sort_values(ascending=False).head(n)
```

```python
def get_stdrpu(df):
    return round(df.groupby('customer_id')['amount'].sum().std(), 2)
def get_stdtpu(df):
    return round(df.groupby('customer_id')['amount'].count().std(), 2)
```

```python
def get_arpu(df):
    return round(df.groupby('customer_id')['amount'].sum().mean(), 2)
def get_atpu(df):
    return round(df.groupby('customer_id')['amount'].count().mean(), 2)
```

```python
def get_customer_pareto(transactions, customers):
    customer_pareto = get_top_n_customers_by_amount(transactions, customers, None)
    pareto = (customer_pareto.cumsum() / customer_pareto.sum()).astype(float)*100
#     pareto.index = [customers.loc[x, 'full_name'] if x in customers.index else x for x in pareto.index]
    return pareto
```

```python
def get_item_pareto(transactions):
    item_pareto = get_top_n_items_by_amount(transactions,None)
    return (item_pareto.cumsum() / item_pareto.sum()).astype(float)*100
```

```python
def get_sendload_allocation(transactions):
    alloc = transactions.groupby('type').sum()['amount']
    return alloc
```

```python
def get_new_customers_on_year_month(customer_reg, year, month):
    return customer_reg[(customer_reg['date_add'].dt.year == year) & (customer_reg['date_add'].dt.month == month)].index
```

```python
def get_last_three_months(year, month):
    if month >=3:
        one_month_ago = (year, month - 1)
        two_months_ago = (year, month - 2)
    else:
        if month == 2:
            one_month_ago = (year, month - 1)
            two_months_ago = (year - 1, 12)
        elif month == 1:
            one_month_ago = (year - 1, 12)
            two_months_ago = (year - 1, 11)
    return [(year, month), one_month_ago, two_months_ago]
```

```python
def get_new_customers_last_3_months(customer_reg, year, month):
    year_months = get_last_three_months(year, month)
    customers = []
    for curr_year, curr_month in year_months:
        customers.extend(list(get_new_customers_on_year_month(customer_reg, curr_year, curr_month)))
    return set(customers)
```

```python
def get_new_customer_sales_ratio(transactions, new_customers):
    overall_sales = get_sales_amount(transactions)
    
    new_customer_transactions = transactions[transactions['shop_id'].isin(new_customers)]
    new_sales = get_sales_amount(new_customer_transactions)
    old_sales = overall_sales - new_sales
    
    return pd.Series({'old': old_sales, 'new': new_sales})
```

```python
def get_new_customer_conversion_rate(transactions, new_customers):
    new_customer_transactions = transactions[transactions['shop_id'].isin(new_customers)]
    transacted = new_customer_transactions['customer_id'].unique()
    
    return len(transacted) / len(new_customers) * 100
```

```python
def get_pct_diff(new, old):
    return 100 * (new - old) / old
```

```python
def data_in_archive(year, month):
    files = [x for x in os.listdir('archive') if x.endswith('csv')]
    for file in files:
        file_year = file.split('_')[1][4:]
        file_month = file.split('_')[2].split('.')[0][5:]
        if year == file_year and month == file_month:
            return True
        else:
            continue
    return False
```

```python
def download_data(year, month):
    if data_in_archive(year, month):
        if datetime.datetime.now().year == year and datetime.datetime.now().month == month:
            shopsend = get_shopsend_transactions(year, month)
            sendload = get_sendload_transactions(year, month)
            overall = pd.concat([shopsend, sendload], ignore_index=True)
        else:
            shopsend = pd.read_csv('archive/shopsend_year{}_month{}.csv'.format(year, month))
            sendload = pd.read_csv('archive/sendload_year{}_month{}.csv'.format(year, month))
            overall = pd.read_csv('archive/overall_year{}_month{}.csv'.format(year, month))
    else:
        shopsend = get_shopsend_transactions(year, month)
        sendload = get_sendload_transactions(year, month)
        overall = pd.concat([shopsend, sendload], ignore_index=True)
    return shopsend, sendload, overall
```

```python
def get_sales_amount_diff(shopsend_current, sendload_current, overall_current, 
                          shopsend_last_month, sendload_last_month, overall_last_month):
    return [get_sales_amount(overall_current), get_pct_diff(get_sales_amount(overall_current), get_sales_amount(overall_last_month))], \
           [get_sales_amount(shopsend_current), get_pct_diff(get_sales_amount(shopsend_current), get_sales_amount(shopsend_last_month))], \
           [get_sales_amount(sendload_current), get_pct_diff(get_sales_amount(sendload_current), get_sales_amount(sendload_last_month))],
```

# number of txns

```python
def get_num_txn_diff(shopsend_current, sendload_current, overall_current, 
                          shopsend_last_month, sendload_last_month, overall_last_month):
    return [get_num_transactions(overall_current), get_pct_diff(get_num_transactions(overall_current), get_num_transactions(overall_last_month))], \
           [get_num_transactions(shopsend_current), get_pct_diff(get_num_transactions(shopsend_current), get_num_transactions(shopsend_last_month))], \
           [get_num_transactions(sendload_current), get_pct_diff(get_num_transactions(sendload_current), get_num_transactions(sendload_last_month))],
```

# arpu

```python
def get_arpu_diff(shopsend_current, sendload_current, overall_current, 
                          shopsend_last_month, sendload_last_month, overall_last_month):
    return [get_arpu(overall_current), get_pct_diff(get_arpu(overall_current), get_arpu(overall_last_month))], \
           [get_arpu(shopsend_current), get_pct_diff(get_arpu(shopsend_current), get_arpu(shopsend_last_month))], \
           [get_arpu(sendload_current), get_pct_diff(get_arpu(sendload_current), get_arpu(sendload_last_month))],
```

# new signups this month

```python
def get_signups_diff(customers, year_current, month_current):
    if month_current == 1:
        year_last = year_current - 1
        month_last = 12
    else:
        year_last = year_current
        month_last = month_current - 1
    return [len(get_new_customers_on_year_month(customers, year_current, month_current)),
            get_pct_diff(len(get_new_customers_on_year_month(customers, year_current, month_current)), 
                 len(get_new_customers_on_year_month(customers, year_last, month_last)))]
```

# conversion rate (wrt new customers in last 3 months)

```python
def get_conversion_diff(customers, transactions_current, transactions_last, year_current, month_current):
    if month_current == 1:
        year_last = year_current - 1
        month_last = 12
    else:
        year_last = year_current
        month_last = month_current - 1
    return [get_new_customer_conversion_rate(transactions_current, get_new_customers_last_3_months(customers, year_current, month_current)),
            get_pct_diff(get_new_customer_conversion_rate(transactions_current, get_new_customers_last_3_months(customers, year_current, month_current)),
                         get_new_customer_conversion_rate(transactions_last, get_new_customers_last_3_months(customers, year_last, month_last)))]
```
